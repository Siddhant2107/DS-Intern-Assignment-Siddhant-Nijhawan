```python
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    return data[(data[column] >= Q1 - 1.5*IQR) & (data[column] <= Q3 + 1.5*IQR)]

# Apply on both
df_clean = remove_outliers_iqr(df, 'random_variable1')
df_clean = remove_outliers_iqr(df_clean, 'random_variable2')
```

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer()
df[['random_variable1', 'random_variable2']] = pt.fit_transform(df[['random_variable1', 'random_variable2']])
```

```python
for col in ['random_variable1', 'random_variable2']:
    df[col].fillna(df[col].median(), inplace=True)
```
